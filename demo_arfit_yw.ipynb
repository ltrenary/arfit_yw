In this notebook, data for an AR(p) model are simulated and plotted. The 'arfit_yw.py' function is used to estimate the parameters of the simulated data.

```python
import warnings
import matplotlib.pyplot as plt
import numpy as np

import statsmodels.api as sm
from statsmodels.tsa.arima_process import ArmaProcess

%run arfit_yw.py 
```

```python
## Construct synthetic AR data with specified parameters ##
arparams = np.array([.95, -.15])
maparams = np.array([1])
ar = np.r_[1, -arparams] # add zero-lag and negate
ma = maparams

AR_object = ArmaProcess(ar, ma)
simulated_data = AR_object.generate_sample(nsample=1000)
plt.plot(simulated_data);
```

```python
## Estimate the coefficients using the arfit_yw function##
max_order = 10
order,phi,aic_ar,res,varp,aicc,aic,bic = arfit_yw(simulated_data,max_order)
```
